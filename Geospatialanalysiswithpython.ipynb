```python
municipality_data.plot(figsize=(10,10))
plt.title('Shapefile Data')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

Color our map using cmap

```python
from matplotlib import colormaps
list(colormaps)
```

```python
municipality_data.plot("PR_NAME",cmap='magma',legend=True)
plt.title('Selection through Province Name')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
municipality_data.plot("DISTRICT",cmap='cividis',legend=True,figsize=(20,20))
plt.title('Selection through District Name')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
ax=municipality_data.plot("PR_NAME",cmap='plasma',figsize=(20,20))
ax.set_axis_off()
```

```python
rupandehi_data.plot("PALIKA",cmap='inferno',legend=True,figsize=(15,15))
```

```python
ax=rupandehi_data.plot("PALIKA",cmap='viridis',figsize=(15,15))
ax.set_axis_off()
```
